```python

import cv2
import numpy as np
import torch
import torch.nn as nn
import os
import timm
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image
def gradcam(weight_path,image_path):
    model = timm.create_model('efficientnetv2_s', pretrained=False)
    in_features = model.classifier.in_features
    model.classifier = nn.Linear(in_features, 3)
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    model_weight_path = weight_path
    assert os.path.exists(model_weight_path), "file {} does not exist.".format(model_weight_path)
    model.load_state_dict(torch.load(model_weight_path, map_location=device))
    model.to(device)
    model.eval()
    target_layers = [model.conv_head]
    # conv_layers = [module for module in model.modules() if isinstance(module, torch.nn.Conv2d)]
    # target_layers = [conv_layers[169]]
    cam = GradCAM(model=model, target_layers=target_layers, use_cuda='cuda:0')
    rgb_img = cv2.imread(image_path, 1)[:, :, ::-1]
    rgb_img = cv2.resize(rgb_img,(384,384))
    rgb_img = np.float32(rgb_img) / 255
    input_tensor= torch.from_numpy(rgb_img)
    input_tensor=input_tensor.permute(2,0,1).unsqueeze(0)
    # print(input_tensor.shape)
    targets = None
    grayscale_cam = cam(input_tensor=input_tensor, targets=targets)
    # In this example grayscale_cam has only one image in the batch:
    grayscale_cam = grayscale_cam[0, :]
    visualization = show_cam_on_image(rgb_img, grayscale_cam, image_weight= 0.85,use_rgb=True)
    # cv2.imwrite(f'visualization.jpg', visualization)
    return visualization
```

```python
weight_path ='/home/yfang/MRI_110_result/1mri_95_efficientv2_s_best_metric_model.pth'
path =r'/home/yfang/dataset/MRI_110'
for i in os.listdir(path):
    for j in os.listdir(os.path.join(path,i)):
            for k in os.listdir(os.path.join(path,i,j)):
                image_path=os.path.join(path,i,j,k)
                
                save_path=os.path.join(r'/home/yfang/gradcam',i,os.path.basename(j))
                if not os.path.exists(save_path):
                    os.makedirs(save_path)
                print(save_path)
                visualization=gradcam(weight_path,image_path)
                cv2.imwrite(os.path.join(save_path,os.path.basename(image_path)) , visualization)   


                
                


```
